```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/simulated_transaction_2024.csv")
```

```python
data.columns
```

```python
display(data.head())
print("=====================================================================================================================")
print(f"The shape of the dataset is {data.shape}.")
print("=====================================================================================================================")
display(data.info())
```

```python
desc = pd.DataFrame(index = list(data))
desc['type'] = data.dtypes
desc['count'] = data.count()
desc['nunique'] = data.nunique()
desc['%unique'] = desc['nunique'] / len(data) * 100
desc['null'] = data.isnull().sum()
desc['%null'] = desc['null'] / len(data) * 100
display(desc)
desc.to_csv("./data/desc.csv")
```

```python
data.describe().drop(['Account No','Third Party Account No'],axis=1).T.to_excel("./data/num_desc.xlsx")
```

```python
data["Account No"] = data["Account No"].apply(lambda x: int(x) if x >= 0 else -1)
data["Third Party Account No"] = data["Third Party Account No"].apply(lambda x: int(x) if x >= 0 else -1)
data["Third Party Name"] = data["Third Party Name"].apply(lambda x: x if type(x) != float else "Unknow")
```

```python
data = data[(data["Third Party Account No"] != -1)| (data["Third Party Name"] != "Unknow")]
display(data.head())
```
